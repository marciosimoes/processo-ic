# Teste 2 - Transformação de Dados

Para este teste, foi usado mais uma vez a biblioteca <code>zipfile</code> para a minipulação dos arquivos zips, <code>read_pdf</code> para manipular o arquivo pdf, <code>io</code> para auxiliar na leitura do pdf em tempo de código e o <code>Pandas</code> para a manipulação dos dados tabelados.

```python
import zipfile
import io
import pandas as pd
from tabula import read_pdf
```

```python
import warnings
warnings.filterwarnings("ignore")
```

Logo, já será feito a leitura do pdf que iremos utilizar (Anexo I) e realizado a captação de todas as tabelas como desejado:

```python
with zipfile.ZipFile('download_pdfs.zip', 'r') as zip_ref:
    pdf_file = zip_ref.open('Anexo_I_Rol_2021RN_465.2021_RN473_RN478_RN480_RN513_RN536_RN537_RN538_RN539_RN541_RN542_RN544_546.pdf')
    df = read_pdf(io.BytesIO(pdf_file.read()), pages='all', multiple_tables=False)
    tabela = df[0]
```

Assim, obtemos a tabela em questão:

```python
tabela.head()
```

Como pedido em seguida faremos a substituição de acordo com a legenda:

```python
replacements = {'OD': 'Seg. Odontológica', 'AMB': 'Seg. Ambulatorial'}
```

```python
tabela.replace(replacements, inplace=True)
tabela.head()
```

E finalmente salvamos em um csv e zipamos com o nome <code>Teste_Marcio.zip</code>

```python
with zipfile.ZipFile('Teste_Marcio.zip', 'w') as zip_file:
    tabela.to_csv(zip_file.open('tabela.csv', 'w'), index=False)
```
